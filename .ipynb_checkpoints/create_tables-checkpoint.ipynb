```python
import numpy as np
import pickle
import pandas as pd

with open('results/objs.pkl', 'rb') as f:
    problems, ratios_run1, ratio_95_conf_ints_run1 = pickle.load(f)
    f.close()

    
with open('results/objs2.pkl', 'rb') as f:
    problems, ratios_run2, ratio_95_conf_ints_run2 = pickle.load(f)
    f.close()

combo_list = [max(performance) for performance in zip(ratios_run1, ratios_run2)]
ratios_table = np.array(combo_list)
ratios_table = np.reshape(ratios_table, (8,6))
ratios_table = np.transpose(ratios_table)
pd.set_option('display.expand_frame_repr', False)
data_tab = pd.DataFrame(ratios_table, ["p = 1/4", "p = 1", "p = 4", "p = 9", "p = 39", "p = 99"],
                        ["L = 1", "L = 4", "L = 10", "L = 20", "L = 30", "L = 50", "L = 70", "L = 100"])
print(" ")
print(data_tab)
print(" ")

best_run = []
for performance in zip(ratios_run1, ratios_run2):
    if performance[1] > performance[0]:
        best_run.append(2)
    else:
        best_run.append(1)

best_conf_int = []
d=0
for j in best_run:
    if j==1:
        best_conf_int.append(ratio_95_conf_ints_run1[d])
    else:
        best_conf_int.append(ratio_95_conf_ints_run2[d])
    d=d+1    

strings = []
for interval in best_conf_int:
    interval[0], interval[1] = round(interval[0]*1000000)/1000000, round(interval[1]*1000000)/1000000 # round to 6 decimal places
    strings.append(str(interval)) # convert entire interval for a problem to a single string and append it to list
    
strings_table = np.array(strings)
strings_table = np.reshape(strings_table, (8,6))
strings_table = np.transpose(strings_table)
data_tab_conf = pd.DataFrame(strings_table, ["p = 1/4", "p = 1", "p = 4", "p = 9", "p = 39", "p = 99"],
                        ["L = 1", "L = 4", "L = 10", "L = 20", "L = 30", "L = 50", "L = 70", "L = 100"])
print(" ")
print(data_tab_conf[["L = 1", "L = 4", "L = 10", "L = 20"]])
print(data_tab_conf[["L = 30", "L = 50", "L = 70", "L = 100"]])
print(" ")


best_run_table = np.array(best_run)
best_run_table = np.reshape(best_run_table, (8,6))
best_run_table = np.transpose(best_run_table)
best_run_table = pd.DataFrame(best_run_table, ["p = 1/4", "p = 1", "p = 4", "p = 9", "p = 39", "p = 99"],
                        ["L = 1", "L = 4", "L = 10", "L = 20", "L = 30", "L = 50", "L = 70", "L = 100"])
print("(Best of the two runs: just for personal reference)")
print(best_run_table)
```
